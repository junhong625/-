#### CSV 파일의 로드(시간으로 파싱)

```python
import pandas as pd
df1 = pd.read_csv('access_log.csv', parse_dates=['time'])
df1
```

#### 시간을 인덱스 지정

```python
df2 = df1.set_index('time')
df2
```

#### 인덱스에 의한 시간 필터링

```python
df3 = df2['1995-07-01' : '1995-07-03']
```

#### 1일분의 액세스 수 카운트

```python
df3.resample('1d').size() 

## 년: y, 월: m, 일:d, 시:h, 분:t, 초:s
```

### ※ 데이터 분석에서는 위와 같이 새로운 변수에 차례대로 값을 대입하면서 데이터를 가공!
